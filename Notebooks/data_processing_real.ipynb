```python
dfs = [precip_01_df, precip_02_df, precip_03_df, precip_04_df, precip_05_df, precip_06_df, 
       precip_07_df, precip_08_df, precip_09_df, precip_10_df, precip_11_df, precip_12_df]
combined_precip_df = pd.concat(dfs, ignore_index=True)
```

```python
##Convert dates to datetime format
argo_float_2020_df['time'] = pd.to_datetime(argo_float_2020_df['time'])
polar_sic_df['time'] = pd.to_datetime(polar_sic_df['time'])
polar_windspeed_df['time'] = pd.to_datetime(polar_windspeed_df['time'])
combined_precip_df['time'] = pd.to_datetime(combined_precip_df['time'])
```

```python
##Group each dataframe by day so that I can split into different dataframes for each day for easier processing
grouped_by_date = argo_float_2020_df.groupby(argo_float_2020_df['time'])
grouped_by_date_sic = polar_sic_df.groupby(polar_sic_df['time'])
grouped_by_date_ws = polar_windspeed_df.groupby(polar_windspeed_df['time'])
grouped_by_date_precip = combined_precip_df.groupby(combined_precip_df['time'])
```

```python
##Split Argo data into dfs for each day
dfs_by_date_argo = {}

for date, group in grouped_by_date:
    key = f"argo_float_{date.strftime('%Y_%m_%d')}"
    dfs_by_date_argo[key] = group
```

```python
##Split sea ice conc data into dfs for each day
dfs_by_date_sic = {}

for date, group in grouped_by_date_sic:
    key = f"sea_ice_conc_{date.strftime('%Y_%m_%d')}"
    dfs_by_date_sic[key] = group
```

```python
##Split windspeed data into dfs for each day
dfs_by_date_ws = {}

for date, group in grouped_by_date_ws:
    key = f"windspeed_{date.strftime('%Y_%m_%d')}"
    dfs_by_date_ws[key] = group
```

```python
##Split precip data into dfs for each day
dfs_by_date_precip = {}

for date, group in grouped_by_date_precip:
    key=f"precip_{date.strftime('%Y_%m_%d')}"
    dfs_by_date_precip[key] = group
```

```python
##Merge dfs in argo dictionary and sea ice conc dictionary 
merged_dfs = {}

#Iterate over the items in the dfs_by_date_argo dictionary
for key, df_argo in dfs_by_date_argo.items():
    #Construct the corresponding key for dfs_by_date_sic
    date_part = key.split('_', 2)[-1]  
    sic_key = f'sea_ice_conc_{date_part}' 
    
    #Check if the corresponding key exists in dfs_by_date_sic
    if sic_key in dfs_by_date_sic:
        df_sic = dfs_by_date_sic[sic_key]
        
        #Merge the data frames
        merged_df = pd.merge(df_argo, df_sic, on=['lat_grid', 'lon_grid'])
        
        merged_key = f'merged_{date_part}'
        
        #Store the merged data frame in the merged_dfs dictionary using the new key
        merged_dfs[merged_key] = merged_df
```

```python
##Merge dfs in merged df dictionary and windspeed dictionary
partially_merged_dfs = {}

#Iterate over the items in the merged_dfs dictionary
for key, merged_df in merged_dfs.items():
    #Construct the corresponding key for dfs_by_date_ws
    date_part = key.split('_', 1)[-1] 
    ws_key = f'windspeed_{date_part}'

    #Check if the corresponding key exists in dfs_by_date_ws
    if ws_key in dfs_by_date_ws:
        df_ws = dfs_by_date_ws[ws_key]

        #Merge the data frames
        partially_merged_df = pd.merge(merged_df, df_ws, on=['lat_grid', 'lon_grid'])

        partially_merged_key = f'partially_merged_{date_part}'

        #Store the merged data frame in the fully_merged_dfs dictionary using the new key
        partially_merged_dfs[partially_merged_key] = partially_merged_df
```

```python
##Merge dfs in partially_merged_dfs dictionary and precip dictionary
fully_merged_dfs = {}

#Iterate over the items in the partially_merged_dfs dictionary
for key, partially_merged_df in partially_merged_dfs.items():
    #Construct the corresponding key for dfs_by_date_precip
    date_part = key.split('_', 2)[-1]
    precip_key = f'precip_{date_part}'

    #Check if the corresponding key exists in dfs_by_date_precip
    if precip_key in dfs_by_date_precip:
        df_precip = dfs_by_date_precip[precip_key]
        #Merge the data frames
        fully_merged_df = pd.merge(partially_merged_df, df_precip, on=['lat_grid', 'lon_grid'], suffixes=('_df1', '_df4'))

        fully_merged_key = f'fully_merged_{date_part}'
        
        #Store the fully merged data frame in the fully_merged_dfs dictionary using the new key
        fully_merged_dfs[fully_merged_key] = fully_merged_df
    
```

```python
##Merging daily dfs by month
dfs_by_ym = {}

# Iterate over the keys in fully_merged_dfs and group them by year-month
for key in fully_merged_dfs:
    year_month = key[13:][:7] 
    
    if year_month not in dfs_by_ym:
        dfs_by_ym[year_month] = [fully_merged_dfs[key]]
    else:
        dfs_by_ym[year_month].append(fully_merged_dfs[key])

# Initialize a dictionary to hold the concatenated data frames by month
concatenated_dfs_by_month = {}

# Concatenate data frames for each year-month group
for ym, dfs in dfs_by_ym.items():
    new_key = f'merged_{ym}'
    concatenated_dfs_by_month[new_key] = pd.concat(dfs)
```

```python
##Concatenate into df for 1 year
merged_2020_df = pd.concat(concatenated_dfs_by_month.values())
```
